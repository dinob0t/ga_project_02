```python
def add_doy(df_in):
    import datetime
    doy_list = []
    for row in df_in.iterrows():
        
        month = row[1]['month']
        year = row[1]['year']
        day = row[1]['day']
        date = datetime.datetime(year,month,day) 
        doy = date.timetuple().tm_yday
        doy_list.append(doy)
                           
    df_in['doy'] = pd.Series(doy_list)
    return df_in

df_A = add_doy(df_A)
df_B = add_doy(df_B)
df_C = add_doy(df_C)
df_D = add_doy(df_D)
df_E = add_doy(df_E)
```

```python
#Remove bad data
def remove_bad_data(df_in):
    temp_thresh = -273.15*10
    pressure_thresh = 0
    df_in = df_in[df_in.temp > temp_thresh]
    df_in = df_in[df_in.pressure > pressure_thresh]
    df_in = df_in[df_in.wind_dir > 0]
    df_in = df_in[df_in.wind_speed > 0]
    
    return df_in

df_A = remove_bad_data(df_A)
df_B = remove_bad_data(df_B)
df_C = remove_bad_data(df_C)
df_D = remove_bad_data(df_D)
df_E = remove_bad_data(df_E)
df_A.tail(30)
```

```python
from sklearn import linear_model
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

# Experimentation:
# Put all our features in: wind_dir is possibly 
# problematic since it wraps around 0->360?
X = df_A[['doy','hour','wind_speed','wind_dir', 'pressure']].values
Y = df_A[['temp']].values

# Had to update anaconda to get latest scikit-learn
# to use polynomialfeatures and pipeline. Had to look up
# how to use these.
# X = np.vander(X[:,0],degree)

# Not exactly sure which classifier / how much complexity
# to build into the model here (bias/variance tradeoff)
# Try Lasso for regularization?
clf = linear_model.Ridge()
degree = 4

polynomial_features = PolynomialFeatures(degree=degree,
                                             include_bias=False)
pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("classifier", clf)])
# Training
pipeline.fit(X, Y[:,0])
```

```python
# Predict all data
predict_Y = pipeline.predict(X)
score = pipeline.score(X,Y)
print score
# R^2 = 0.85, 0.86, 0.84, 0.85, 0.82 seems OK for cities A,B,C,D,E.
# Should really look at other metrics too

# Plot model against data
plt.scatter(X[:,0],Y)
plt.plot(X[:,0],predict_Y)

plt.ylabel('Temp')
plt.xlabel('Doy')
plt.show()
```

```python
def generate_classifiers(city_data):
    X = city_data[['doy','hour','wind_speed','wind_dir', 'pressure']].values
    Y = city_data[['temp']].values

    clf = linear_model.Ridge()
    degree = 4

    polynomial_features = PolynomialFeatures(degree=degree,
                                                 include_bias=False)
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                             ("classifier", clf)])
    pipeline.fit(X, Y[:,0])
    return pipeline
```

```python
clf_cityA = generate_classifiers(df_A)
clf_cityB = generate_classifiers(df_B)
clf_cityC = generate_classifiers(df_C)
clf_cityD = generate_classifiers(df_D)
clf_cityE = generate_classifiers(df_E)
```

```python
# To predict the temperature on Jan 1st, 2012 we can
# use the pressure and wind data from the previous day
# and use day of the year (doy)=1

# Have to correct some missing data
XA = df_A.tail(21)
XB = df_B.tail(20)
XC = df_C.tail(15)
XD = df_D.tail(22)
XE = df_E.tail(19)

# Replace missing values with last values - should use interpolation
def interp_missing_values(data_in,doy):  
    last_data = 0
    for hour in range(0,24): 
        row = data_in[data_in.hour == hour]
        row.doy = doy
        if hour == 0:
            data_out = row[['doy','hour','wind_speed','wind_dir', 'pressure']].values
        # If missing value, replace with last
        elif row.empty:
            last_row = data_in[data_in.hour == last_data]
            last_row.hour = hour
            last_row.doy = doy
            data_out = np.concatenate((data_out,last_row[['doy','hour','wind_speed','wind_dir', 'pressure']].values))
        else:
            data_out = np.concatenate((data_out,row[['doy','hour','wind_speed','wind_dir', 'pressure']].values))
            last_data = hour
        if len(data_out)>23:
            break
    return data_out
XA = interp_missing_values(XA,1) 
XB = interp_missing_values(XB,1) 
XC = interp_missing_values(XC,1) 
XD = interp_missing_values(XD,1) 
XE = interp_missing_values(XE,1) 
```

```python
predict_YA = clf_cityA.predict(XA)
predict_YB = clf_cityA.predict(XB)
predict_YC = clf_cityA.predict(XC)
predict_YD = clf_cityA.predict(XD)
predict_YE = clf_cityA.predict(XE)
```

```python
# To predict March 3rd, 2012 we can use average temperatures from March 3rd on previous years
# since we have no recent data

def get_average_temps(month, day, data_in):
    count_dict = {x:0 for x in range(0,24)}
    temp_dict = {x:0 for x in range(0,24)}
    for year in range(data_in.head(1).year,data_in.tail(1).year + 1):
        df = data_in[(data_in.year == year) & (data_in.month == month) & (data_in.day == day)]
        for hour in range(0,24):
            cur_row = df[df.hour == hour]
            if not cur_row.empty:
                count_dict[hour] = count_dict[hour] + 1
                temp_dict[hour] = temp_dict[hour] + int(cur_row.temp)
    # Make list of average temps
    temp_list = []
    for hour in range(0,24):
        temp_list.append(temp_dict[hour]/count_dict[hour])
    return temp_list
                
        
av_tempA = get_average_temps(3,3,df_A)
av_tempB = get_average_temps(3,3,df_B)
av_tempC = get_average_temps(3,3,df_C)
av_tempD = get_average_temps(3,3,df_D)
av_tempE = get_average_temps(3,3,df_E)
```
