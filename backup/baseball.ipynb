```python
def add_ywc(df_in):
    count = 0
    ywc_list = np.zeros([df_in.shape[0]+1])
    
    for row in df_in.iterrows():
        cur_playerID = row[1].playerID
        cur_team = row[1].teamID
        cur_year = row[1].yearID
        year_count = 1
        prev_year_team = df[(df.yearID == cur_year-year_count) & (df.playerID == cur_playerID)].teamID.values       
        while len(prev_year_team)>0:
            if prev_year_team[0] != cur_team:
                break
            year_count += 1
            prev_year_team = df[(df.yearID == cur_year-year_count) & (df.playerID == cur_playerID)].teamID.values       
        ywc_list[count] = year_count
        count +=1
    df_in['ywc'] = pd.Series(ywc_list)
    return df_in
```

```python
df = add_ywc(df)
```

```python
from sklearn.cross_validation import cross_val_score, LeaveOneOut
from scipy.stats import sem
from sklearn import metrics
def loo_cv(X_train,y_train,clf):
    # Perform Leave-One-Out cross validation
    loo = LeaveOneOut(X_train[:].shape[0])
    scores=np.zeros(X_train[:].shape[0])
    for train_index,test_index in loo:
        X_train_cv, X_test_cv= X_train[train_index], X_train[test_index]
        y_train_cv, y_test_cv= y_train[train_index], y_train[test_index]
        clf = clf.fit(X_train_cv,y_train_cv)
        y_pred=clf.predict(X_test_cv)
        scores[test_index]=metrics.accuracy_score(np.exp(y_test_cv.astype(int)), np.exp(y_pred.astype(int)))
    print ("Mean score: {0:.3f} (+/-{1:.3f})").format(np.mean(scores), sem(scores))
```

```python
from sklearn import linear_model
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.cross_validation import train_test_split

# Experimentation:
df_test = df[df.teamID == 'HOU']

# X = df_test[['yearID','switched','G','G_batting','AB','R','H','X2B','X3B','HR','RBI','SB','CS','BB','SO','IBB','HBP','SH','SF','GIDP']].values
X = df_test[['yearID','age','G_batting','G_pitching','ywc','R_cum']].values
y = df_test[['log_salary']].values

# y_train = y_train[:,0]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=33)

# Had to update anaconda to get latest scikit-learn
# to use polynomialfeatures and pipeline. Had to look up
# how to use these.
# X = np.vander(X[:,0],degree)

# Not exactly sure which classifier / how much complexity
# to build into the model here (bias/variance tradeoff)
# Try Lasso for regularization?
clf = linear_model.LinearRegression()
degree = 3

polynomial_features = PolynomialFeatures(degree=degree,
                                             include_bias=False)
pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("classifier", clf)])
# Training
pipeline.fit(X_train, y_train)
print X_train.shape, y_train.shape, X.shape, y.shape, X_test.shape, y_test.shape
```
