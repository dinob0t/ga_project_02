```python
def add_pitching(df_in):
    df_pitching = pd.read_csv('Pitching.csv')
    count = 0
    # Add new blank columns for all added pitching data
    zero_list = np.zeros([df_in.shape[0]]) 
    pitching_features = ['W', 'L', 'G', 'GS', 'CG', 'SHO', 'SV', 'IPouts', 'H', 'ER', 'HR', 'BB', 'SO', 'BAOpp', 'ERA', 'IBB', 'WP', 'HBP', 'BK', 'BFP', 'GF', 'R', 'SH', 'SF', 'GIDP']
    for feature in pitching_features:
        new_str = 'p'+feature
        df_in[new_str] = zero_list
        
    for row in df_in.iterrows():     
        cur_playerID = row[1].playerID
        cur_year = row[1].yearID
        row_lookup = (df_pitching.playerID == cur_playerID) & (df_pitching.yearID == cur_year)
        for feature in pitching_features:
            cur_val = df_pitching.loc[row_lookup,feature].values
            if len(cur_val)>0:
                cur_val = cur_val[0]
            else:
                cur_val = 0.0
            new_str = 'p'+feature
            df_in.ix[count,new_str] = cur_val
        count +=1
    return df_in
```

```python
if fresh ==1:
    df = add_pitching(df)
    # Replace HOFID with binary 0 no, 1 yes
    df['hofID'].fillna(0.0,inplace=True)
    df.loc[df['hofID'] != 0.0,'hofID'] = 1.0
    # Replace nickname with binary
    df['nameNick'].fillna(0.0,inplace=True)
    df.loc[df['nameNick'] != 0.0,'nameNick'] = 1.0
    # Replace throws and bats L = 0, R = 1
    df.loc[df['bats'] =='L','bats'] = 0.0
    df.loc[df['bats'] =='R','bats'] = 1.0
    df.loc[df['bats'] =='B','bats'] = 2.0
    df.loc[df['throws'] =='L','throws'] = 0.0
    df.loc[df['throws'] =='R','throws'] = 1.0
    # Delete weird data
    df = df[df.salary > 1]
    df = df[df.yearID > 1]
    df = df[df.weight > 2]
    df = df[df.height > 2]
    df = df.drop(['debut','finalGame','managerID','deathYear','deathMonth','deathCountry','deathState','deathDate','deathDay', 'lahman40ID', 'lahman45ID', 'retroID', 'holtzID','deathCity','nameFirst','nameLast','nameNote'],axis=1)

    # Set NaN stats to zero
    for feature in df.columns:
        df[feature].fillna(0.0,inplace=True)
    
    # Add in age
    df['age'] = df['yearID']-df['birthYear']
    # Add in batting ratio -> 0 pitchers, 1 batters
    df['Batting_ratio']=df.G_batting/df.G
    # Sort for cumulative stats below
    df = df.sort(columns=['playerID','yearID'])
    df.index = range(0,len(df))
```

```python
def add_ywc(df_in):
    count = 0
    df_in['ywc'] = np.zeros([df_in.shape[0]])
    for row in df_in.iterrows():
        if count == 0:
            df_in[count,'ywc'] = 1
            prev_playerID = row[1].playerID
            prev_team = row[1].teamID
            prev_year = row[1].yearID
        else:            
            cur_playerID = row[1].playerID
            cur_team = row[1].teamID
            cur_year = row[1].yearID
            if cur_playerID == prev_playerID and cur_team == prev_team and cur_year == prev_year+1:
                df_in.ix[count,'ywc'] = df_in.ix[count-1,'ywc'] + 1
                prev_year = cur_year
            else:
                df_in.ix[count,'ywc'] = 1
                prev_playerID = cur_playerID
                prev_team = cur_team
                prev_year = cur_year
        
        count += 1
    return df_in
```

```python
if fresh == 1:
    df = add_ywc(df)
```

```python
def add_cum_stats(df_in):
    
    zero_list = np.zeros([df_in.shape[0]]) 
    features = ['G', 'G_batting', 'AB', 'R', 'H', 'X2B', 'X3B', 'HR', 'RBI', 'SB', 'CS', 'BB', 'SO', 'IBB', 'HBP', 'SH', 'SF', 'GIDP','pW', 'pL', 'pG', 'pGS', 'pCG', 'pSHO', 'pSV', 'pIPouts', 'pH', 'pER', 'pHR', 'pBB', 'pSO', 'pBAOpp', 'pERA', 'pIBB', 'pWP', 'pHBP', 'pBK', 'pBFP', 'pGF', 'pR', 'pSH', 'pSF', 'pGIDP']
    for feature in features:
        new_str = feature + '_cum'
        df_in[new_str] = zero_list

    count = 0
    cur_playerID = ''   
    for row in df_in.iterrows(): 
        if row[1].playerID == cur_playerID:
            
            for feature in features:
                new_str = feature + '_cum'
                df_in.ix[count,new_str] = df_in.ix[count-1,new_str] + df_in.ix[count,feature] 
        else:
            for feature in features:
                new_str = feature + '_cum'
                df_in.ix[count,new_str] = df_in.ix[count,feature] 
            cur_playerID = row[1].playerID
        count += 1
        if count > 30:
            break
    df_in['Batting_ratio_cum']=df_in.G_batting_cum/df_in.G_cum    
    return df_in
```

```python
if fresh == 1:
    df = add_cum_stats(df)
```

```python
def log_stats(df_in):
    # Take log of some data
    df_in['log_ywc'] = np.log(df_in.ywc)
    df_in['log_salary'] = np.log(df_in.salary)
    # Use logs for cumulative stats
    features = ['G', 'G_batting', 'AB', 'R', 'H', 'X2B', 'X3B', 'HR', 'RBI', 'SB', 'CS', 'BB', 'SO', 'IBB', 'HBP', 'SH', 'SF', 'GIDP','pW', 'pL', 'pG', 'pGS', 'pCG', 'pSHO', 'pSV', 'pIPouts', 'pH', 'pER', 'pHR', 'pBB', 'pSO', 'pBAOpp', 'pERA', 'pIBB', 'pWP', 'pHBP', 'pBK', 'pBFP', 'pGF', 'pR', 'pSH', 'pSF', 'pGIDP']        
    for feature in features:
        cum_feature = feature + '_cum'
        new_str = 'log_' + cum_feature
        df_in[new_str] = df_in[cum_feature]
        select_gt_zero = df_in[cum_feature]>0
        df_in.loc[select_gt_zero,new_str] = np.log(df_in.loc[select_gt_zero,cum_feature])
    return df_in
```

```python
if fresh == 1:
    df = log_stats(df)
```

```python
from sklearn import linear_model
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
# from sklearn import svm

for feature in df.columns:
    if feature != 'salary' and feature != 'log_salary':

        df_test = df
    #     df_test = df_test.dropna(subset=features,how='any')    
    #     df_test.index = range(0,len(df_test))

        X = df_test[feature].values
        y = df_test[['log_salary']].values

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=33)

        clf = linear_model.LinearRegression()
        degree = 3

        polynomial_features = PolynomialFeatures(degree=degree,
                                                     include_bias=False)
        pipeline = Pipeline([("polynomial_features", polynomial_features),
                                 ("classifier", clf)])
    #     pipeline = svm.SVR(kernel='poly', degree=2)
        pipeline.fit(X_train, y_train)

        predict_y = pipeline.predict(X)

        cur_rsq = pipeline.score(X,y)
        cur_mse = mean_squared_error(predict_y, y)
        cur_mae = mean_absolute_error(predict_y, y)

        print 'Summary: '
        print 'Features: ', feature
        print 'R^2: ', (cur_rsq)
        print 'MSE: ', (cur_mse)
        print 'MAE: ', (cur_mae)
```
