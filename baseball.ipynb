```python
def add_pitching(df_in):
    df_pitching = pd.read_csv('Pitching.csv')
    count = 0
    ERA_list = np.zeros([df_in.shape[0]+1])
    SO_list = np.zeros([df_in.shape[0]+1])
    IPouts_list = np.zeros([df_in.shape[0]+1])
    for row in df_in.iterrows():     
        cur_playerID = row[1].playerID
        cur_year = row[1].yearID
        pitching_row = df_pitching[(df_pitching.playerID == cur_playerID) & (df_pitching.yearID == cur_year)]
#         cur_ERA = df_pitching[(df_pitching.playerID == cur_playerID) & (df_pitching.yearID == cur_year)].ERA.values
#         cur_SO = df_pitching[(df_pitching.playerID == cur_playerID) & (df_pitching.yearID == cur_year)].SO.values
        cur_ERA = pitching_row.ERA.values
        cur_SO = pitching_row.SO.values
        cur_IPouts = pitching_row.IPouts.values
        if len(cur_ERA)>0:
            ERA_list[count] = cur_ERA[0]
        else:
            ERA_list[count] = 0
        if len(cur_SO)>0:
            SO_list[count] = cur_SO[0]
        else:
            SO_list[count] = 0
        if len(cur_IPouts)>0:
            IPouts_list[count] = cur_IPouts[0]
        else:
            IPouts_list[count] = 0
        count +=1
    df_in['pERA'] = pd.Series(ERA_list)
    df_in['pSO'] = pd.Series(SO_list)
    df_in['pIPouts'] = pd.Series(IPouts_list)
    return df_in
```

```python
def add_ywc(df_in):
    count = 0
    ywc_list = np.zeros([df_in.shape[0]+1])
    
    for row in df_in.iterrows():
        cur_playerID = row[1].playerID
        cur_team = row[1].teamID
        cur_year = row[1].yearID
        year_count = 1
        prev_year_team = df[(df.yearID == cur_year-year_count) & (df.playerID == cur_playerID)].teamID.values       
        while len(prev_year_team)>0:
            if prev_year_team[0] != cur_team:
                break
            year_count += 1
            prev_year_team = df[(df.yearID == cur_year-year_count) & (df.playerID == cur_playerID)].teamID.values       
        ywc_list[count] = year_count
        count +=1
    df_in['ywc'] = pd.Series(ywc_list)
    return df_in
```
